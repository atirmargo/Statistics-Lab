```python
import random
import pandas as pd
import numpy as np
from scipy import stats
import math
```

1. Найти медиану, выборочное среднее, выборочную дисперсию.

```python
n = 100
X = [random.gauss(0, math.sqrt(2)) for _ in range(n)]
mean = stats.tmean(X)        # Выборочное среднее
median = stats.tmean(X)  # Медиана
variance = stats.tvar(X)  # Выборочная дисперсия 
print("Результаты для выборки из N(0, 2):")
print(f"Размер выборки: {n}")
print(f"Выборочное среднее: {mean}")
print(f"Медиана: {median}")
print(f"Выборочная дисперсия: {variance}")
```

Найти mean sin X

```python
n = 100
X = np.array([random.gauss(0, math.sqrt(2)) for _ in range(n)])
sin_X = np.sin(X)
mean_sinX = np.mean(sin_X)
print(f"Выборочное среднее sinX: {mean_sinX}")
#[sin(x) for x in X]
```

Извлечь из этой выборки новую выборку, содержащую только положительные значения.

```python
import numpy as np
n = 100
X = np.random.normal(0, np.sqrt(2), n)
positive_X = X[X>0]
print(f"Положительная значения: {positive_X}")
```

Пусть Sk = X1 + · · · + Xk. Найти минимальное k такое, что Sk > 3. Если такого нет, вывести n + 1.

```python
n = 100
X = np.random.normal(0, np.sqrt(2), n)
current_sum = 0
min_k = n + 1  #по дефолту
for k in range(n):
    current_sum += X[k]
    if current_sum > 3:
        min_k = k + 1  
        break
print(f"Минимальное k такое, что Sk > 3: {min_k}")
```

3.4. Найти число студентов с баллом ЕГЭ по математике, лежащим в заданном диапазоне

```python
min = 80
max = 90
filtered_students = (df['ЕГЭ...матем.'] >= min) & (df['ЕГЭ...матем.'] <= max)
print(filtered_students.sum())
```

3.5. Создать датасет с количествами студентов, баллы которых лежат в заданных диапазонах (<70,
70-80 и т.д.)

```python
bins_detailed = [0, 70, 80, 90, 95, float('inf')]
labels_detailed = ['<70', '70-80', '80-90', '90-95', '95+']

df['Диапазон_баллов'] = pd.cut(df['ЕГЭ...матем.'], bins=bins_detailed, labels=labels_detailed, right=False)
result_detailed = df.groupby('Диапазон_баллов').agg(Количество=('ЕГЭ...матем.', 'count')).reset_index()
result_detailed = result_detailed.sort_values('Диапазон_баллов')

print(result_detailed)
```
